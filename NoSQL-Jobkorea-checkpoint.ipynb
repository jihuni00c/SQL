```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time

URL = "https://www.jobkorea.co.kr/"

service = Service(ChromeDriverManager().install())
options = Options()

options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--lang=ko_KR")

driver = webdriver.Chrome(service=service, options=options)
driver.get(URL)
time.sleep(4)

# 검색
search = driver.find_element(By.CSS_SELECTOR, "input.block.border-0.text-size-b3-15.text-weight-regular.text-color-gray950")
search.clear()
search.send_keys("마케팅")
search.send_keys(Keys.ENTER)
time.sleep(3)




# 검색 결과 공고 카드 가져오기
job_posts = driver.find_elements(By.CSS_SELECTOR, "div[data-sentry-element='Box']")

results = []

for post in job_posts:
    try:
        title = post.find_element(By.CSS_SELECTOR, ".styles_mb_space2__dk46ts4t").text.strip()
        company = post.find_element(By.CSS_SELECTOR, ".Flex_display_inline-flex__i0l0hl1.Flex_gap_space6__i0l0hl1g.Flex_align_center__i0l0hl8.styles_mb_space20__dk46ts2h").text.strip()
        location = post.find_element(By.CSS_SELECTOR, ".Flex_display_flex__i0l0hl2.Flex_gap_space4__i0l0hly.Flex_align_center__i0l0hl8.styles_flex-shrink_0__dk46tsa9.styles_px_space10__dk46tsm._10fvqwx0").text.strip().split("외")[0].strip()
        keyword = post.find_element(By.CSS_SELECTOR, ".Flex_display_flex__i0l0hl2.Flex_gap_space4__i0l0hly.Flex_align_center__i0l0hl8.styles_flex-shrink_1__dk46tsaa.styles_px_space10__dk46tsm._10fvqwx0").text.strip()
        
        results.append({
            "모집명": title,
            "모집기관": company,
            "지역": location,
            "키워드": keyword
        })
        
    except:
        continue


print(results)

driver.quit()
```
